```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 의사결정 나무 분류 모델
from sklearn.tree import DecisionTreeClassifier
```

```python
#학습용, 평가용 데이터 구분하기 위한 패키지 불러오기
from sklearn.model_selection import train_test_split
```

```python
#데이터 불러오기
path = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
raw = pd.read_csv(path)

df = raw.copy()

df.head(3)
```

```python
df.shape
#891개 데이터, 12개의 속성
```

```python
df.info()
#데이터 결측치(not-null), 데이터 유형
```

```python
# 요약 통계량
df.describe()
```

```python
df.columns
```

```python
# 결측치 확인 및 개수 세기
df['Age'].isnull().sum()
```

```python
# 단순대치법 : 평균
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace=True)
```

```python
df.info()
```

```python
df['Embarked'].value_counts()
```

```python
# 단순대치법 - 가장 빈도수가 많은 최빈값을 사용

df['Embarked'].mode()[0]
```

```python
embarked_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(embarked_mode, inplace=True)
```

```python
df.info()
```

```python
df['Sex']
```

```python
# Sex 컬럼의 값을 1과 0으로 라벨 인코딩 함
from sklearn.preprocessing import LabelEncoder

df['Sex'] = LabelEncoder().fit_transform(df['Sex'])
```

```python
df['Sex'].head()
```

```python
df['Embarked']
```

```python
df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df['Embarked']
```

```python
df.columns
```

```python
# 기존 변수 활용 새로운 변수(파생변수) 생성
df['family'] = df['Parch'] + df['SibSp']

df.head()
```

```python
df.columns
```

```python
# 분석 데이터 준비
# x : 독립변수, y : 종속변수

x = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'family']]
y = df['Survived']
```

```python
print(x)
print()
print(y)
```

```python
# 데이터 분할 (train_data, test_data)
x_train, x_test, y_train, y_test =\
train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)

# 반드시 훈련은 누구에게? 훈련용 데이터
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
# 모델 성능 비교 (정확도 측정)
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
acc
```

```python
# KNN(K 최근접 이웃) 모델

from sklearn.neighbors import KNeighborsClassifier
```

```python
# 훈련 끝난 모델 >> 예측
y_pred = knn.predict(x_test)
y_pred
```

```python
y_test
```
