```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#의사결정나무
from sklearn.tree import DecisionTreeClassifier

# 학습용 데이터와 테스트 데이터(평가용 데이터, 검증용 데이터) 분리
from sklearn.model_selection import train_test_split

# path 다음에 깃허브 csv 주소 입력해도 가능.
path = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
```

```python
# 원본을 살려놓기 위해 카피한 후 사용하는 법
raw = pd.read_csv(path)

df = raw.copy()
```

```python
df.head(3)
```

```python
print(df.shape) # 행렬 150개 데이터, 5개 열(속성)
print()
print(df.info())
print()
print(df.describe())
```

```python
# 텍스트로 되어 있는 sepcies 데이터 >> 숫자 0, 1, 2 로 변환

df['species'].replace({'setosa':0, 'versicolor':1, 'virginica':2}, inplace=True)

# df['species'] = df['species'].replace({'setosa':0, 'versicolor':1, 'virginica':2})
# in place of 명사/동명사 : ~을 대신하다
```

```python
# 분석 데이터 준비
# x : 독립변수(설명변수, 원인변수)
# y : 종속변수(결과변수)

df.columns
```

```python
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
x
```

```python
y = df['species']
y
```

```python
x_train, x_test, y_train, y_test =\
train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델링
# 의사결정나무 객체 생성

dt = DecisionTreeClassifier(random_state=42)

# 학습 수행(최적화 작업(정규화 작업 포함)
# >> 모델 구축(훈련용 데이터만 사용))
dt.fit(x_train, y_train)
```

```python
# 학습완료 >> 테스트 데이터로 확인

y_pred = dt.predict(x_test)
```

```python
# 모델 성능 비교 : 정확도

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred) # 비교대상 : 실제값, 모델이 예측한 값
print(acc)
```

```python
# 분석 모델 성능 평가 방법

# 모델 성능 - 오차 행렬(confusion matrix)
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
# 모델 성능 평가 - 평가지표 계산
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)

print(report)
```
